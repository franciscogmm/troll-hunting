```python
db = load_database(conn, 'posts')
db.reload()
posters_collection = load_collection(db, 'Posters', document=1)
reactors_collection = load_collection(db, 'Reactors', document=1)
posts_collection = load_collection(db, 'Posts', document=1)
edge_collection = load_collection(db, 'Edges', document=0)
```

## Add Post to Posts Collection

```python
list_of_post_nodes = create_list_of_node_stores(posts_collection)
result = search_for_existing_node(
    list_of_post_nodes,
    search_param='original_request_url',
    search_term=post_important_only['original_request_url'],
    verbose=0,
)
post_doc = assess_node_search_result(result, posts_collection, post_important_only)
```

## Add Posters to Posters Collection

```python
poster_docs = []

list_of_poster_nodes = create_list_of_node_stores(posters_collection)
try:
    retrieved_post_doc = post_doc.getStore()
except:
    retrieved_post_doc = deepcopy(post_doc)

print(retrieved_post_doc['username'])
poster_profile = {
    'user_id': retrieved_post_doc['user_id'],
    'username': retrieved_post_doc['username'],
    'link': f"https://facebook.com/profile.php?id={retrieved_post_doc['user_id']}&fref=pb",
}

result = search_for_existing_node(
    list_of_poster_nodes, search_param='user_id', search_term=retrieved_post_doc['user_id'], verbose=0
)
poster_doc = assess_node_search_result(result, posters_collection, poster_profile)
poster_docs.append(poster_doc)
print(f'Saved {len(poster_docs)} records')
```

## Add Potential Trolls to Reactors Collection

```python
reactor_docs = []

list_of_reactor_nodes = create_list_of_node_stores(reactors_collection)
for troll in neg_reactors:
    print(troll['name'])
    troll_profile = {
        'name': troll['name'],
        'link': troll['link'],
    }

    result = search_for_existing_node(list_of_reactor_nodes, search_param='link', search_term=troll['link'], verbose=0)
    troll_doc = assess_node_search_result(result, reactors_collection, troll_profile)
    reactor_docs.append(troll_doc)

print(f'Saved {len(reactor_docs)} records')
```

## Create Poster Edges

```python
from_collection_name = 'Posters'
to_collection_name = 'Posts'

list_of_edges = create_list_of_node_stores(edge_collection)
for poster in poster_docs:
    try:
        link = poster.getStore()['link']
        user_id = poster.getStore()['user_id']
        username = poster.getStore()['username']
    except:
        link = poster['link']
        user_id = poster['user_id']
        username = poster['username']

    dict_edge = {'edge_type': 'posted'}

    print(f'Poster {link}/ {username} / {poster["_key"]} {dict_edge["edge_type"]} post {post_doc["_key"]}')

    result = search_for_existing_edge(
        list_of_edges,
        search_param_from='_from',
        search_term_from=f'{from_collection_name}/{poster["_key"]}',
        search_param_to='_to',
        search_term_to=f'{to_collection_name}/{post_doc["_key"]}',
        search_param_addtl='edge_type',
        search_term_addtl=dict_edge['edge_type'],
        verbose=0,
    )

    edge = assess_edge_search_result(
        result,
        edge_collection=edge_collection,
        from_collection_name=from_collection_name,
        to_collection_name=to_collection_name,
        from_node_doc=poster,
        to_node_doc=post_doc,
        dict_edge=dict_edge,
    )
```

## Create Reactor Edges

```python
from_collection_name = 'Reactors'
to_collection_name = 'Posts'

list_of_reaction_edge = create_list_of_node_stores(edge_collection)
for reactor in reactor_docs:
    try:
        link = reactor.getStore()['link']
    except:
        link = reactor['link']

    dict_edge = {'edge_type': 'reacted', 'reaction_type': [x['type'] for x in neg_reactors if x['link'] == link][0]}

    print(f'Reactor {link}/{reactor["_key"]} reacts {dict_edge["reaction_type"]} to post {post_doc["_key"]}')

    result = search_for_existing_edge(
        list_of_reaction_edge,
        search_param_from='_from',
        search_term_from=f'{from_collection_name}/{reactor["_key"]}',
        search_param_to='_to',
        search_term_to=f'{to_collection_name}/{post_doc["_key"]}',
        search_param_addtl='reaction_type',
        search_term_addtl=dict_edge['reaction_type'],
        verbose=0,
    )

    edge = assess_edge_search_result(
        result,
        edge_collection=edge_collection,
        from_collection_name=from_collection_name,
        to_collection_name=to_collection_name,
        from_node_doc=reactor,
        to_node_doc=post_doc,
        dict_edge=dict_edge,
    )
```
